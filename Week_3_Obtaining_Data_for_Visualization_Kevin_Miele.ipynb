```python
import numpy as np


# Cover the domain from -2π to 2π
x = np.linspace(-2*np.pi, 2*np.pi, 1000)

# Compute the sine and cosine values
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot sin(x) with a red solid line and cos(x) with a green dashed line
plt.plot(x, y_sin, color='red', label='sin(x)')
plt.plot(x, y_cos, color='green',  label='cos(x)')

# Label the axes and add a title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plotting Sine and Cosine Curves')

# Add a legend
plt.legend(loc='lower right')

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from -2π to 2π
data = np.random.normal(0, 1, 1000)

# Label axes and add title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Plot of sin(x) and cos(x)')

# Plot histogram and change color
plt.hist(data,bins=90, color = 'red')
plt.show()

# Calculate the mean of the data array
mean_value = np.mean(data)
# Draw a vertical line at the mean value with a contrasting color
plt.axvline(mean_value, color='red', label='Mean')
```

Answer Reflection Questions:

How does changing the number of bins affect the histogram's appearance and interpretation?:

**When you increase the number of bins, for example: from 30 to 60, the histogram expands horizontally and gets thinner. The opposite occurs if you go lower as it shrinks and gets thicker. **

What can you infer about the distribution of data from the histogram and the mean line?

**Data of large quantities, like population and world data are commonly associated with histograms as they provide a overall consensus with an easy visual representation. Histograms are also easy to associate with means and other data concepts like standard deviation. **
