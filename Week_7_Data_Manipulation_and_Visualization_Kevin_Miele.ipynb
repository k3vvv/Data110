2. Data Filtering

```python
import seaborn as sns

# Load the penguins dataset
penguins_df = sns.load_dataset('penguins')

# Filter the dataset to include only penguins with body mass greater than 4000 grams
penguins_greater_than_4000g = penguins_df[penguins_df['body_mass_g'] > 4000]

# Display the first 5 rows of the filtered dataset
print("First 5 rows of the filtered dataset:")
print(penguins_greater_than_4000g.head())
```

3. Conditional Filtering

```python
import seaborn as sns

# Load the penguins dataset
penguins_df = sns.load_dataset('penguins')

# Create a subset of the data based on the given conditions
subset_df = penguins_df[(penguins_df['bill_length_mm'] >= 40) &
                        (penguins_df['bill_length_mm'] <= 50) &
                        (penguins_df['bill_depth_mm'] < 18)]

# Display the shape of the resulting DataFrame
print("Shape of the resulting DataFrame:", subset_df.shape)
```

4. Data Grouping and Aggregation

```python
import seaborn as sns

# Load the penguins dataset
penguins_df = sns.load_dataset('penguins')

# Group the dataset by species and island, and calculate the average flipper length for each group
avg_flipper_length_by_group = penguins_df.groupby(['species', 'island'])['flipper_length_mm'].mean()

# Sort the results in descending order of flipper length
sorted_avg_flipper_length = avg_flipper_length_by_group.sort_values(ascending=False)

# Display the sorted results
print("Average flipper length (mm) for each species and island (sorted in descending order):")
print(sorted_avg_flipper_length)
```

5. Advanced Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the penguins dataset
penguins_df = sns.load_dataset('penguins')

# Set up the plot
plt.figure(figsize=(10, 6))

# Create histogram with different colors for each species
sns.histplot(data=penguins_df, x='body_mass_g', hue='species', palette='muted', multiple='stack', bins=15)

# Add title and labels
plt.title('Distribution of Penguin Body Masses by Species')
plt.xlabel('Body Mass (g)')
plt.ylabel('Frequency')

# Add legend with explicit labels
plt.legend(title='Species', labels=penguins_df['species'].unique(), loc='upper right')

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

6. Utilizing Groups for Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the penguins dataset
penguins_df = sns.load_dataset('penguins')

# Get unique island names
islands = penguins_df['island'].unique()

# Set up subplots
fig, axes = plt.subplots(nrows=1, ncols=len(islands), figsize=(15, 5), sharey=True)

# Loop through each island and create a histogram
for island, ax in zip(islands, axes):
    # Filter data for the current island
    island_data = penguins_df[penguins_df['island'] == island]
    # Create histogram for flipper lengths
    sns.histplot(data=island_data, x='flipper_length_mm', ax=ax, bins=15)
    # Set title for the histogram
    ax.set_title(f'Distribution of Flipper Lengths\n({island.capitalize()} Island)')
    # Set labels for axes
    ax.set_xlabel('Flipper Length (mm)')
    ax.set_ylabel('Frequency')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
