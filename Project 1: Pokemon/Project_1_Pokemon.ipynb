5 Most Common Types of Pokemon:

```python
import pandas as pd

# Data from my Github
url = 'https://raw.githubusercontent.com/k3vvv/Data110/d3e27fd2cdfca659bd9983781861fd1e054ba91e/pokemon.csv'
pokemon_data = pd.read_csv(url)

# Count the frequency of each type in Type 1
type1_counts = pokemon_data['Type1'].value_counts()

# Count the frequency of each type in Type 2
type2_counts = pokemon_data['Type2'].value_counts()

# Combine the counts from both columns
total_counts = type1_counts.add(type2_counts, fill_value=0)

# Sort the counts in descending order
sorted_counts = total_counts.sort_values(ascending=False)

# Display the top 5 most common types
print("Top 5 most common types of Pokémon:")
print(sorted_counts.head())
```

Bar Graph format:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data from my Github
url = 'https://raw.githubusercontent.com/k3vvv/Data110/d3e27fd2cdfca659bd9983781861fd1e054ba91e/pokemon.csv'
pokemon_data = pd.read_csv(url)

# Count the frequency of each type in Type 1
type1_counts = pokemon_data['Type1'].value_counts()

# Count the frequency of each type in Type 2
type2_counts = pokemon_data['Type2'].value_counts()

# Combine counts from both columns
total_counts = type1_counts.add(type2_counts, fill_value=0)

# Sort the counts from highest to lowest
sorted_counts = total_counts.sort_values(ascending=False)

# The number of top types to display (Top selection option: top N)
top_n = 5

# Get the top N (Top selection option) most common types
top_types = sorted_counts.head(top_n)

# Create a bar chart
plt.figure(figsize=(10, 6))
top_types.plot(kind='bar', color='Red')
plt.title('Top 5 Most Common Types of Pokémon')
plt.xlabel('Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Heatmap for the different regions:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dataset from my Github
url = 'https://raw.githubusercontent.com/k3vvv/Data110/d3e27fd2cdfca659bd9983781861fd1e054ba91e/pokemon.csv'
pokemon_data = pd.read_csv(url)

# Code to fix error revolving around missing Trainer Region Values (Remember to keep the zeros in the heatmap to avoid more errors)
pokemon_data = pokemon_data.dropna(subset=['Trainer Region'])

# Count the number of captures for each Trainer Region
region_captures = pokemon_data['Trainer Region'].value_counts().reset_index()
region_captures.columns = ['Trainer Region', 'Captures']

# Rearrange the columns and rows to create a format that is easy for creating a heatmap
# Create the heatmap
heatmap_data = region_captures.pivot(index='Trainer Region', columns='Captures', values='Captures').fillna(0)
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, cmap='Blues', annot=True, fmt='g', linewidths=0.5)
plt.title('Pokémon Captures by Trainer Region')
plt.xlabel('Captures')
plt.ylabel('Trainer Region')
plt.show()
```

Scatterplot for different Pokemon Types (Type1) and (Type2):

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dataset from my GitHub
url = 'https://raw.githubusercontent.com/k3vvv/Data110/d3e27fd2cdfca659bd9983781861fd1e054ba91e/pokemon.csv'
pokemon_data = pd.read_csv(url)

# Matching each Pokémon type to a unique color (Try to be as accurate as possible and try to use bright colors)
type_color_mapping = {
    'Normal': 'skyblue', 'Fire': 'orangered', 'Water': 'royalblue',
    'Electric': 'yellow', 'Grass': 'limegreen', 'Ice': 'lightcyan',
    'Fighting': 'firebrick', 'Poison': 'mediumorchid', 'Ground': 'sandybrown',
    'Flying': 'lightskyblue', 'Psychic': 'mediumpurple', 'Bug': 'olive',
    'Rock': 'saddlebrown', 'Ghost': 'darkviolet', 'Dragon': 'mediumslateblue',
    'Dark': 'dimgray', 'Steel': 'lightsteelblue', 'Fairy': 'pink', 'Unknown': 'gray'
}

# Create a scatter plot for Type1 and Type2 combinations with different colors for each type
plt.figure(figsize=(12, 8))
sns.scatterplot(data=pokemon_data, x='Type1', y='Type2', hue='Type1', palette=type_color_mapping, s=100)
plt.title('Scatter Plot of Pokémon Type Combinations')
plt.xlabel('Type1')
plt.ylabel('Type2')
plt.xticks(rotation=45)
plt.legend(title='Type1', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
