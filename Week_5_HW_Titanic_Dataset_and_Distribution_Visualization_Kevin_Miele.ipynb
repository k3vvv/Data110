2. Exploratory Data Analysis (EDA)
Perform an initial analysis to understand the survival rate:

Calculate and compare the overall survival rate to the survival rates by gender.
Explore the age distribution of the dataset.
Analyze survival rates by passenger class.

```python
import pandas as pd
import seaborn as sns

# Loading the Titanic dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/71e2436a092d714350de0fc409ca8a8714e7e78f/titanic.csv"
titanic_df = pd.read_csv(url)

# Calculate the overall survival rate
overall_survival_rate = titanic_df['survived'].mean()
print("Overall Survival Rate:", overall_survival_rate)

# Compare survival rates by gender
survival_rate_by_gender = titanic_df.groupby('sex')['survived'].mean()
print("\nSurvival Rates by Gender:")
print(survival_rate_by_gender)

# Explore the age distribution of the dataset
sns.histplot(data=titanic_df, x='age', kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Analyze survival rates by passenger class
survival_rate_by_class = titanic_df.groupby('class')['survived'].mean()
print("\nSurvival Rates by Passenger Class:")
print(survival_rate_by_class)
```

3.1 Histograms for Age Distribution
Create histograms to compare the age distributions of survivors and non-survivors.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create histograms for age distribution of survivors and non-survivors
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic_df, x='age', hue='survived', kde=True, multiple='stack')
plt.title('Age Distribution of Survivors and Non-Survivors')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend(labels=['Non-Survivors', 'Survivors'])
plt.show()
```

3.2 Pyramid-style Age Distribution
Construct a pyramid-style plot to compare age distributions between survivors and non-survivors. This visualization will help highlight which age groups were more likely to survive.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the Titanic dataset
titanic_df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/71e2436a092d714350de0fc409ca8a8714e7e78f/titanic.csv")

# Separate data for survivors and non-survivors
survived = titanic_df[titanic_df['survived'] == 'yes']
non_survived = titanic_df[titanic_df['survived'] == 'no']

# Set up the figure and subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Plot the histogram for survivors
sns.histplot(data=survived, x='age', ax=axes[0], bins=20, color='skyblue', kde=True)
axes[0].set_title('Survived')
axes[0].set_ylabel('Count')

# Plot the histogram for non-survivors, inverted
sns.histplot(data=non_survived, x='age', ax=axes[1], bins=20, color='salmon', kde=True)
axes[1].set_title('Non-Survived')
axes[1].invert_xaxis()  # Invert the x-axis to create the pyramid effect

# Set common labels and display the plot
plt.xlabel('Age')
plt.tight_layout()
plt.show()
```

3.3 KDE Plots for Survival by Age ( for male, female and total)
KDE Plot for Survival by Age: Visualize the age distribution for survivors and non-survivors using KDE plots. This will provide a smooth estimate of the age distributions.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the Titanic dataset
titanic_df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/71e2436a092d714350de0fc409ca8a8714e7e78f/titanic.csv")

# Set up the figure and subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# KDE Plot for Survival by Age (Total)
sns.kdeplot(data=titanic_df, x='age', hue='survived', ax=axes[0], fill=True)
axes[0].set_title('Total')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Density')

# KDE Plot for Survival by Age (Male)
sns.kdeplot(data=titanic_df[titanic_df['sex'] == 'male'], x='age', hue='survived', ax=axes[1], fill=True)
axes[1].set_title('Male')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('')

# KDE Plot for Survival by Age (Female)
sns.kdeplot(data=titanic_df[titanic_df['sex'] == 'female'], x='age', hue='survived', ax=axes[2], fill=True)
axes[2].set_title('Female')
axes[2].set_xlabel('Age')
axes[2].set_ylabel('')

# Adjust layout
plt.tight_layout()
plt.show()
```

4.2 Bar Graph of Death by Class
Construct a bar graph to visualize the number of deaths by passenger class, either side by side or stacked, to examine the relationship between socio-economic status and survival.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic_df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/71e2436a092d714350de0fc409ca8a8714e7e78f/titanic.csv")

# Create the bar plot
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic_df, x='class', hue='survived', palette='Set1')
plt.title('Number of Deaths by Passenger Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

Based on the visualizations and analysis of the Titanic dataset, several factors appear to have influenced survival rates:

Gender: One of the most prominent factors influencing survival rates was gender. Women had a significantly higher chance of survival compared to men. This trend can be seen from the bar graph comparing survival rates by gender, where the number of female survivors is higher than male survivors.

Age: The age distribution of survivors and non-survivors suggests that children and young adults had higher chances of survival. The KDE plots for survival by age show that the age distribution of survivors is skewed towards younger ages.

Passenger Class: Passenger class also played a role in determining survival. The bar graph of passenger class distribution reveals that the majority of passengers belonged to the third class, followed by the first and second classes. However, when examining the number of deaths by passenger class, it becomes apparent that passengers in the first class had a higher chance of survival compared to those in the second and third classes. This suggests that socio-economic status had an impact on survival rates, with passengers in higher classes having better access to lifeboats and other resources.
